```python
def selection(arr):
    arr.sort(key = lambda x: x[1])
    dynamic = [1] * len(arr)
    i = 1
    maximum = arr[0][1]
    sub = 1
    while(i < len(arr)):
        if arr[i][0] > maximum:
            dynamic[i] = sub + 1
            sub = dynamic[i]
            maximum = arr[i][1]
        else:
            result = list()
            for j in range(0,i):
                if arr[i][0] > arr[j][1]:
                    result.append(dynamic[j])
                dynamic[i] = max(result, default = 0) + 1
        
        i += 1
    return dynamic
```

```python
re = selection(activity)
```
