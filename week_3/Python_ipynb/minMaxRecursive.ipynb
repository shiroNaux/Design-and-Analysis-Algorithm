```python
from numpy.random import seed 
from numpy.random import randint 
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import time

def findMin(A, n): 
    if (n == 1): 
        return A[0] 
    return min(A[n - 1], findMin(A, n - 1))

def findMax(A,n):
    if (n == 1): 
        return A[0] 
    return max(A[n - 1], findMax(A, n - 1))

number = list() 
times_min = list() 
times_max = list()

for i in range (1,11):
    t = 0
    for j in range(1,31):
        a = randint(0, 200 * i, 200 * i)
        start = time.process_time()
        findMin(a, len(a))
        end = time.process_time()
        t += end - start
    
    number.append(len(a))
    times_min.append(t/30)
    
for i in range (1,11):
    t = 0
    for j in range(1,31):
        a = randint(0, 200 * i, 200 * i)
        start = time.process_time()
        findMax(a, len(a))
        end = time.process_time()
        t += end - start
    times_max.append(t/30)  
    
figure(num = None, figsize=(8, 4), dpi=100, facecolor='w', edgecolor='k') 

plt.subplot(1, 2, 1)
plt.xlabel('Số phần tử của mảng')
plt.ylabel('Thời gian chạy')
plt.plot(number, times_min, 'ro')
plt.title('Tìm min')
plt.grid()
plt.tight_layout()

plt.subplot(1, 2, 2)
plt.xlabel('Số phần tử của mảng')
plt.ylabel('Thời gian chạy')
plt.title('Tìm max')
plt.plot(number, times_max, 'bo')
plt.grid()
plt.tight_layout()
```
